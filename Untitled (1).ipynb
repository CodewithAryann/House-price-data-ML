## Load the Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score
```

```python
# Load the data
data = pd.read_csv('House_price.csv')  # or use the sample data from the previous message

print(data)
```

## Clean the Data

```python
# Drop non-useful columns
data.drop(columns=['ID'], inplace=True)

# Convert categorical to numerical
data['HasGarage'] = data['HasGarage'].map({'Yes': 1, 'No': 0})

# Convert YearBuilt to HouseAge
data['HouseAge'] = 2025 - data['YearBuilt']
data.drop(columns='YearBuilt', inplace=True)

# One-hot encode Location
data = pd.get_dummies(data, columns=['Location'], drop_first=True)

# Check for missing values
print(data.isnull().sum())
```

## Feature Scaling

```python
scaler = StandardScaler()
scaled_cols = ['Size_sqft', 'HouseAge', 'Distance_to_CityCenter_km']
data[scaled_cols] = scaler.fit_transform(data[scaled_cols])
```

## Prepare Features and Target

```python
X = data.drop(columns='Price')
y = data['Price']
```

## Train-Test Split

```python
model = LinearRegression()
scores = cross_val_score(model, X, y, cv=3, scoring='r2')  # 3-fold cross-validation

print("Cross-Validated R² Scores:", scores)
print("Average R² Score:", scores.mean())
```

## Fit Model and View Coefficients

```python
model.fit(X, y)

# Show model coefficients
coefficients = pd.Series(model.coef_, index=X.columns)
print("\nFeature Coefficients:\n", coefficients)
```

## Visualize Feature Importance

```python
coefficients.sort_values().plot(kind='barh', figsize=(10, 6), color='skyblue')
plt.title('Feature Importance on House Price')
plt.axvline(x=0, color='gray', linestyle='--')
plt.xlabel('Effect on Price')
plt.tight_layout()
plt.show()
```
